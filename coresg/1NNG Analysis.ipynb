```python
#Run NNDescent for each dataset, get 1NNG and compare edges with MST
for number, dsname in files.items():
    #Open file
    dataset = pd.read_csv(dsname, header=None).values
    #Run pyNNDescent
    index = pynndescent.NNDescent(dataset, n_neighbors = k, verbose=True)
    #extracting 1NNG
    oneNN_mst_edges = [ (min(i, n[1]), max(i, n[1])) for i, n  in enumerate(index.neighbor_graph[0])]
    oneNN_mst_edges = set(oneNN_mst_edges)
    #Printing edges
    print(f"Number of 1NN edges = {len(oneNN_mst_edges)}/{len(dataset)-1}\n")
```
